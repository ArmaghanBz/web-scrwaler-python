```python
#Task 1

mList = [2,3,4,4,6,3,4,8,6,9,1,2]

def highFrequency(mList):
  return max(set(mList), key = mList.count)

print("The number with highest frequency is: ",highFrequency(mList))
```

```python
#Task 2

number = int(input('Please enter The number: '))

def isPalindrome(n):
  if str(n) == str(n)[::-1]:
    print("The number is Palindrome")
  else:
    print("The number is not Palindrome")

isPalindrome(number)


```
